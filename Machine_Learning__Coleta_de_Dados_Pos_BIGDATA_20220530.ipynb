##**Captura IBGE - Pesquisa Nacional de Saneamento Básico - 2017**

```
folder_path = '/content/drive/MyDrive/Pos/SANEAMENTO 2017/'

Saneamento_10 = pd.read_csv(folder_path + "Pesquisa Nacional de Saneamento Básico (2017)-AC.csv", sep=',', encoding="UTF-8")
Saneamento_10 = Saneamento_10.iloc[0:0,:].copy() 

lista_de_arquivos = glob.glob(folder_path +'*.csv')

for nome_do_arquivo in lista_de_arquivos:
    Saneamento_05 = pd.read_csv(nome_do_arquivo, sep=',', encoding="UTF-8")
    Saneamento_05 ['UF']=str(nome_do_arquivo[-6:-4])
    Saneamento_10 = pd.concat([Saneamento_10, Saneamento_05], axis=0, sort=False)

query = ['1.1.1', '1.2.1']

Saneamento_20 = Saneamento_10[Saneamento_10['Posição'].isin(query)]

colunas = ['Nome', 'Localidade', 'UF', '2017']
Saneamento_20 = Saneamento_20 [colunas].copy()

Saneamento_2017 = Saneamento_20.pivot_table(index=['Localidade', 'UF'], columns= "Nome", values= '2017', aggfunc='sum')

Saneamento_2017 = Saneamento_2017.reset_index()

Saneamento_2017.columns = ['Localidade', 'UF', 'Saneamento_Edificios_Abastecidos_Agua_2017', 'Saneamento_Edificios_Esgotados_2017']

Saneamento_2017 ['Saneamento_Edificios_Abastecidos_Agua_2017'] = Saneamento_2017 ['Saneamento_Edificios_Abastecidos_Agua_2017'].replace('-', 0).copy()
Saneamento_2017 ['Saneamento_Edificios_Esgotados_2017'] = Saneamento_2017 ['Saneamento_Edificios_Esgotados_2017'].replace('-', 0).copy()
Saneamento_2017 ['Saneamento_Edificios_Abastecidos_Agua_2017'] = Saneamento_2017 ['Saneamento_Edificios_Abastecidos_Agua_2017'].fillna(0).copy()
Saneamento_2017 ['Saneamento_Edificios_Esgotados_2017'] = Saneamento_2017 ['Saneamento_Edificios_Esgotados_2017'].fillna(0).copy()
Saneamento_2017 ['Saneamento_Edificios_Abastecidos_Agua_2017'] = Saneamento_2017 ['Saneamento_Edificios_Abastecidos_Agua_2017'].astype('float64')
Saneamento_2017 ['Saneamento_Edificios_Esgotados_2017'] = Saneamento_2017 ['Saneamento_Edificios_Esgotados_2017'].astype('float64')

Saneamento_2017.head()
```

##**Captura IBGE - Panorama municipal**

```
folder_path = '/content/drive/MyDrive/Pos/Panorama municipal/'

Panorama_10 = pd.read_csv(folder_path + "Panorama municipal-AC.csv", sep=',', encoding="UTF-8")
Panorama_10 = Panorama_10.iloc[0:0,:].copy() 

lista_de_arquivos = glob.glob(folder_path +'*.csv')

for nome_do_arquivo in lista_de_arquivos:
    Panorama_05 = pd.read_csv(nome_do_arquivo, sep=',', encoding="UTF-8")
    Panorama_05 ['UF']=str(nome_do_arquivo[-6:-4])
    Panorama_10 = pd.concat([Panorama_10, Panorama_05], axis=0, sort=False)

query = ['1.2', '1.3', '3.1.1', '3.1.2', '4.1', '4.2.1', '5.1', '5.2.1.1.1', '5.3']

Panorama_20 = Panorama_10[Panorama_10['Posição'].isin(query)]

dic = {"3.1.2": "Anos_finais_do_ensino_fundamental_2015", "1.2": "Esgotamento_sanitário_adequado_%_2010", 
       "1.3": "Urbanização_de_vias_públicas_%_2010", "3.1.1": "Anos_iniciais_do_ensino_fundamental_2015",
       "4.1": "Internações_por_diarreia_2016", "4.2.1": "Mortalidade_infantil_2012_2014",
       "5.1": "População_ocupada_2019", "5.2.1.1.1": "Classe_de_rendimento_nominal_mensal_domiciliar_per_capita_Até_meio_salário_mínimo_%_2010",
       "5.3": "Salário_médio_mensal_2015"}

dic = {r"\b{}\b".format(k): v for k, v in dic.items()}

Panorama_20["Posição"] = Panorama_20["Posição"].replace(dic, regex=True)

colunas = ['Posição', 'Localidade', 'UF', '2010', '2012-2014', '2014', '2015', '2016', '2017', '2018', '2019']

Panorama_20 = Panorama_20 [colunas].copy()

Panorama_20 = Panorama_20.fillna(0).copy()

Panorama_20 ['2018'] = Panorama_20 ['2018'].astype('float64')

for column in Panorama_20:
    if Panorama_20[column].sum() == 0:
        Panorama_20.drop([column], axis=1, inplace=True)

Panorama_Municipal_2010_2019 = Panorama_20.pivot_table(index=['Localidade', 'UF'], columns= "Posição", values= ['2010', '2012-2014', '2014', '2015', '2016', '2017', '2018', '2019'], aggfunc='sum')

Panorama_Municipal_2010_2019 = Panorama_Municipal_2010_2019.reset_index()

for column in Panorama_Municipal_2010_2019:
    if Panorama_Municipal_2010_2019[column].sum() == 0:
        Panorama_Municipal_2010_2019.drop([column], axis=1, inplace=True)

Panorama_Municipal_2010_2019.drop([('2015', 'População_ocupada_2019'), ('2014', 'Salário_médio_mensal_2015'), ('2014', 'População_ocupada_2019'), ('2016', 'População_ocupada_2019'), ('2017', 'População_ocupada_2019'), ('2018', 'População_ocupada_2019')], axis=1, inplace=True)

Panorama_Municipal_2010_2019.columns = Panorama_Municipal_2010_2019.columns.droplevel(0)

Panorama_Municipal_2010_2019.columns = ['Localidade', 'UF', 'Rendimento_mensal_per_capita_Até_meio_SM_%_2010',
       'Saneamento_Esgotamento_adequado_%_2010',
       'Urbanização_%_vias_públicas_2010', 'Mortalidade_infantil_PER_MIL_2012_2014',
       'Educacao_Anos_finais_Fundamental_2015',
       'Educacao_Anos_iniciais_Fundamental_2015', 'Rendimento_médio_mensal_SM_2015',
       'Internações_Diarreia_PER_MIL_2016', 'População_ocupada_%_2019']

Panorama_Municipal_2010_2019 = Panorama_Municipal_2010_2019.fillna(0).copy()

Panorama_Municipal_2010_2019.head()
```
